<h4 align="center">Comment on the datatype of variables

```python
# the info method of pandas dataframe gives detailed information about the columns and everything and hence using it.
crx_data.info()
```

```python
for i in crx_data:
    print(type(i))
```

```python
# as the above code displayed the names of column it prints column's name type which was string
for i in crx_data:
    print(crx_data[i].dtype)
```

```python
float_,int_,object_=[],[],[]
for i in crx_data:
    if crx_data[i].dtype=="object":
        object_.append(i)
    elif crx_data[i].dtype == "float64":
        float_.append(i)
    elif crx_data[i].dtype == "int64":
        int_.append(i)
    else:
        print(i,"something went wrong")
```

```python
print("there are", len(float_) ,"columns having datatype as float and they are :- ",end=" ")
for i in float_:
    print(i,end=" , ")
print("\nthere are", len(int_) ,"columns having datatype as int and they are :- ",end=" ")
for i in int_:
    print(i,end=" , ")
print("\nthere are", len(object_) ,"columns having datatype as string and they are :- ",end=" ")
for i in object_:
    print(i,end=" , ")
```
