<h4 align="center">Dividing the data into 2 parts i.e Train and test using scikit learn

```python
from sklearn.model_selection import train_test_split
```

<h4 align="center"> making variables into binary format using pd.cut

```python
# cut method is used to segment and sort data values into bins 
# it also convert continous data to categorical data
xtrain_binarised = xtrain.apply(pd.cut, bins=2, labels =[1,0])
xtest_binarised = xtest.apply(pd.cut, bins=2, labels =[1,0])
print("xtrain_binarised type :- ",type(xtrain_binarised))
print("xtest_binarised type:- ",type(xtest_binarised))
```

```python
# converting the binarised values to array
xtrain_binarised = xtrain_binarised.values
xtest_binarised = xtest_binarised.values
print('xtrain_binarised :- ',type(xtrain_binarised))
print('xtest_binarised  :- ',type(xtest_binarised))
```

```python
i=100
max_accuracy=[]
for b in range(xtrain.shape[1]+1):
    accurate_rows= 0
    for x,y in zip(xtrain_binarised,ytrain):
        ypred=(np.sum(x)>=b)
        accurate_rows = accurate_rows + (y == ypred)
    print("For b = ",str(b)+"\n"+"Number of accurate rows",str(accurate_rows),"Accuracy Score :- ", str((accurate_rows/xtrain_binarised.shape[0]))+"\n\n")
    max_accuracy.append([accurate_rows,(accurate_rows/xtrain_binarised.shape[0])])
```

```python
yarraytest=ytest.values
```

```python
i=100
max_accuracy_test=[]
for b in range(xtest.shape[1]+1):
    accurate_rows= 0
    for x,y in zip(xtest_binarised,yarraytest):
        ypred=(np.sum(x)>=b)
        accurate_rows += (y == ypred)
    print("For b = ",str(b)+"\n"+"Number of accurate rows",str(accurate_rows),"Accuracy Score :- ", str((accurate_rows/xtrain_binarised.shape[0]))+"\n\n")
    max_accuracy_test.append([accurate_rows,(accurate_rows/xtrain_binarised.shape[0])])
```

```python
# Using for loop to find maximum accuracy
max_row_train,max_acc_train=0,0
for i in max_accuracy:
    if i[1]>max_acc_train:
        max_acc_train = i[1] 
        max_row_train=i[0]
    else:
        pass
```

```python
# Using for loop to find maximum accuracy
max_row_test,max_acc_test=0,0
for i in max_accuracy_test:
    if i[1]>max_acc_test:
        max_acc_test = i[1] 
        max_row_test=i[0]
    else:
        pass
```
